```python
def Enc(filename, key):
    file = open(filename, "rb")
    data = file.read()
    file.close()
    
    
    
    data = bytearray(data)
    for index, value in enumerate(data):
        data[index] = value ^ key
        
    
    file = open("CC-" + filename, "wb")
    file.write(data)
    file.close()
    
def Dect(filename, key):
    file = open(filename, "rb")
    data = file.read()
    file.close()
    
    data = bytearray(data)
    for index, value in enumerate(data):
        data[index] = value ^ key
        
    
    file = open(filename, "wb")
    file.write(data)
    file.close()
    


atmpt = ""
while atmpt != "3":
    print("pls select opt.")
    print("1. Encrypt File")
    print("2. Decrypt File")
    print("3. Quit")
    atmpt = input()
    if atmpt == "1" or atmpt == "2":
        key = int(input("Enter a key as int!\n"))
        filename = input("Enter filename with extension:\n")
    if atmpt == "1":
        Enc(filename, key)
    if atmpt == "2":
        Dect(filename, key)
```
